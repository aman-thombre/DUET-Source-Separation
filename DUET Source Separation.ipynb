# DUET Source Separation

The DUET Source Separation algorithm uses recordings of the same audio from two different microphones at slightly different locatiosn to separate out sources. This notebook explores how this algorithm works, and tests it out on some synthetic toy test cases.

```python
plt.figure(figsize=(15,6))
plt.plot(wav0,label="Mic 0")
plt.plot(wav1,'--',label="Mic 1")
plt.legend()
plt.title("Mic 0 and Mic 1")
plt.xlabel("Sample")
plt.ylabel("Level")
```

In this test case for dataset 3, we can see the three sources very distinctly. In a more realistic case, the three sources would be mixed, but this provides for a nice visualization. If we zoom in on each test case, we can see that these recordings are delayed versions of one another, with each source having a slightly different delay between microphones. This corresponds to the different distances each source is from each microphone.

```python
plt.figure(figsize=(15,30))

# Source 1
plt.subplot(311)
plt.plot(np.arange(5000,5050),wav0[5000:5050], label="Mic 0")
plt.plot(np.arange(5000,5050),wav1[5000:5050], '--', label="Mic 1")
plt.legend()
plt.title("Zoomed in Audio Recordings for Source 1")
plt.xlabel("Sample")
plt.ylabel("Level")

# Source 2
plt.subplot(312)
plt.plot(np.arange(25000,25050),wav0[25000:25050], label="Mic 0")
plt.plot(np.arange(25000,25050),wav1[25000:25050], '--', label="Mic 1")
plt.legend()
plt.title("Zoomed in Audio Recordings for Source 2")
plt.xlabel("Sample")
plt.ylabel("Level")

# Source 3
plt.subplot(313)
plt.plot(np.arange(50000,50050),wav0[50000:50050], label="Mic 0")
plt.plot(np.arange(50000,50050),wav1[50000:50050], '--', label="Mic 1")
plt.legend()
plt.title("Zoomed in Audio Recordings for Source 3")
plt.xlabel("Sample")
plt.ylabel("Level")
```

From these graphs we can see that:
* For source 1, mic 1 lags behind mic 0 by about 2 samples.
* For source 2, mic 1 is ahead of mic 0 by about 2 samples.
* For source 3, mic 1 and mic 0 are in sync.

The DUET source separation algorithm uses these different delays to distinguish source 1, source 2, and source 3. It'll take the recording from one of the microphones (say mic 0) as the reference, and notice that different audio components in mic 1's recording are delayed by different amounts. The algorithm assumes that each source will have its own delay, as each source is a certain distance from the microphones. The algorithm will then group audio components with similar delays together to separate each source. 

In a nutshell, it should say:
* Every audio component where mic 1 lags behind mic 0 by about 2 samples belongs to source 1.
* Every audio component where mic 1 is ahead of mic 0 by about 2 samples belongs source 2.
* Every audio component where mic 1 and mic 0 have no delay belongs to source 3.

### Implementation

Now, we'll take a look at the implementation of the DUET algorithm, and understand how it works.

First, to take advantage of the delays between recordings saw above, we notice that these audio recordings are simply out of phase with one another. So, we want to extract the phase shift between these audio recordings, and group audio components based on the phase shift. The first step of doing this is taking a short-time Fourier transform (STFT):

*Note that the write-up is for dataset3, but you can change the dataset to visualize the other ones as well.*

```python
# Helper function so we don't take a log of 0
def safe_log(x):
    y = np.zeros(x.shape)
    y[x > np.exp(-100)] = np.log(x[x > np.exp(-100)])
    return y
```

```python
dataset="dataset3" # change this
NUM_SOURCES = 3 # change this

file0 = os.path.join(dataset, "0.wav")
srate,wav0 = scipy.io.wavfile.read(file0)
file1 = os.path.join(dataset, "1.wav")
srate,wav1 = scipy.io.wavfile.read(file1)

# Get STFTs
f0,t0,Zxx0 = signal.stft(wav0,nperseg=2047)
f1,t1,Zxx1 = signal.stft(wav1,nperseg=2047)

# Plot STFTs
plt.figure(figsize=(15,6))
plt.subplot(121)
plt.pcolormesh(t0, f0, safe_log(np.abs(Zxx0)), cmap='viridis',shading='nearest')
plt.title("Spectrogram of Mic 0 Recording")

plt.subplot(122)
plt.pcolormesh(t1, f1, safe_log(np.abs(Zxx1)), cmap='viridis',shading='nearest')
plt.title("Spectrogram of Mic 1 Recording")
```

To get the phase shift of these STFTs, we can simply divide one by the other, and retain the imaginary part of the STFTs: $$\frac{e^{j\omega_{0}}}{e^{j\omega_{1}}} = e^{j(\omega_{0}-\omega_{1})}.$$

```python
# Need to normalize by the frequencies we use when we divide, so grab a matrix of the frequencies between (-pi, pi)
freq = ((np.concatenate((np.arange(1,513),np.arange(-512,0))))*((2*3.14159265)/1024)).reshape(-1,1)
fmat = np.ones((Zxx1.shape[1],freq.shape[0])).T
fmat *= freq

# Get delays
ratio = (Zxx0+1e-32)/(Zxx1+1e-32) # get ratio of STFT so we can get delay
delay = (np.imag(np.log(ratio))/fmat).reshape(-1,1) # get delay from imaginary
```

Let's take a look at the histogram of delays:

```python
delay_plot = delay.flatten()
m = np.mean(delay_plot)
s = np.std(delay_plot)
delay_plot = delay_plot[delay_plot < m+7*s] # Get rid of outliers
delay_plot = delay_plot[delay_plot > m-7*s]

counts,bins = np.histogram(delay_plot,bins=50)
plt.figure(figsize=(15,6))
plt.hist(bins[:-1],bins,weights=counts)
plt.title("Histogram of delays")
plt.xlabel("Delay, in terms of phase shift")
plt.ylabel("Counts")
plt.show()
```

As expected, we can see three very distinct peaks - this means we can use a clustering algorithm to classify each component to a peak. We can use sklearn's KMeans clustering to do this: 

```python
kmeans = KMeans(n_clusters=NUM_SOURCES) # cluster object
kmeans.fit_predict(delay) # clustering
labels = kmeans.labels_.reshape(ratio.shape) # put our labels back into the right shape
```

We can visualize the labels we've assigned on top of our original waveform:

```python
plt.figure(figsize=(15,12))
plt.subplot(211)
plt.pcolormesh(t1, f1, labels, cmap='viridis',shading='nearest')

plt.subplot(212)
plt.plot(wav0)
plt.show()
```

Our original audio sources were pretty distinct, so this is an easy way to visualize that we separated them out pretty well. If there were mixing, we'd have different assignments along the frequency axis as well. Putting it all together, we can isolate our audio and take inverse STFTs to get back into the time domain:

```python
f = Zxx0 * (labels==0) # isolate sources
t,x = signal.istft(f) # inverse transform
source1 = signal.savgol_filter(x,11,2)[:wav0.shape[0]] # smoothed output, same length as original due to 0-padding
```

```python
f = Zxx0 * (labels==1) # isolate sources
t,x = signal.istft(f) # inverse transform
source2 = signal.savgol_filter(x,11,2)[:wav0.shape[0]] # smoothed output, same length as original due to 0-padding
```

```python
f = Zxx0 * (labels==2) # isolate sources
t,x = signal.istft(f) # inverse transform
source3 = signal.savgol_filter(x,11,2)[:wav0.shape[0]] # smoothed output, same length as original due to 0-padding
```

```python
source1 = source1.astype(np.int16)
source2 = source2.astype(np.int16)
source3 = source3.astype(np.int16)
```
